# Estimating the value of $\pi$ using a Monte Carlo technique 

The program picks random points in the area $[-1,1] \times [-1,1]$, and determines which fraction of these are within the unit circle. This in turn gives a measure of $\pi$ with associated statistical uncertainty. Performing such "experiments" many times not only gives the value of $\pi$, but also a feel for the use of Monte Carlo, and experience in calculating averages, RMSs, and the error on these.
The first part of the program is mostly for illustration, while the exercise is question 4 on using this in higher dimension problems.

```python
import numpy as np                                     # Matlab like syntax for linear algebra and functions
import matplotlib.pyplot as plt                        # Plots and figures like you know them from Matlab
from iminuit import Minuit                             # The actual fitting tool, better than scipy's
import sys                                             # Modules to see files and folders in directories
from scipy import stats

plt.style.use('classic')
```

and define the parameters of the program:

```python
save_plots = True         # Determining if plots are saved or not
r = np.random
r.seed(42)

# Set parameters:
N_experiments = 100       # Number of "experiments" determining pi
N_points      = 2000      # Number of points per experiment in determining pi
N_dim = 4                 # Number of dimensions

pi_true = np.pi
```

```python
#calculate the expected precision of 1 monte carlo simulation of pi with 2000 points 
expected_frac = pi_true/4
expected_error = 4 * np.sqrt(expected_frac*(1-expected_frac)/N_points)      # the binomial error is normalized, so we need to multiply by the area of the square
print('expected uncertainty on the fraction of hits inside the circle: ', expected_error)
```

## Loop over process:

```python
pi_estimate    = np.zeros(N_experiments)


for iexp in range(N_experiments):

    uniform_numbers = r.uniform(-1, 1, (N_dim, N_points))            # Generate uniform points between -1 and 1 in N_dim dimensions:
    hit_dist = np.sqrt(np.sum(uniform_numbers**2, axis=0))           # Calculate the distance of each point from the origin:
    Nhit = np.sum(hit_dist < 1, axis=0)                              # Calculate the number of points within the circle:

    

    f  = Nhit / N_points                                             #fraction of points within the circle                           
    sf = np.sqrt(f*(1-f)/N_points)                                   #binomial error on the fraction


    
    # From this we can get pi and its error, and save it:
    pi_estm  =  (f)**(1/np.floor(N_dim / 2))
    pi_error =  (sf)**(1/np.floor(N_dim / 2))
    pi_estimate[iexp] = pi_estm

    # Calculate the integration constant:
    int_const = pi_estm / pi_true
    pi_estm = pi_estm / int_const

    # Print first couple of pi measurements:
    if (iexp < 5) :
        print(f"{iexp:2d}. pi estimate:   {pi_estm:7.4f} +- {pi_error:6.4f}")
        print('fraction:', f)
        print('Integration constant:', int_const)
```

```python
if (N_experiments > 1):
    pi_mean = pi_estimate.mean()
    pi_std = pi_estimate.std(ddof=1)
    pi_standard_error = pi_std / np.sqrt(N_experiments)
    int_const_mean = 1 / (pi_mean * (1 - pi_mean))
    print(f"The {N_experiments:d} experiments yield: pi = {pi_mean:7.5f} +- {pi_standard_error:7.5f}     (std = {pi_std:5.3f})")
    print(f"and the agreement with the true value of pi is: {(pi_mean - pi_true) / pi_standard_error:6.2f} sigma")
else :
    print("Since only one experiment was done, no average and RMS can be calculated.")
```

## Plot the histograms:

Distribution of points from one experiment:

```python
fig1, ax1 = plt.subplots(figsize=(8, 8))    

# Make mask of points inside vs outside the circle:
mask_inside = hit_dist < 1
mask_outside = ~mask_inside

# Make scatter plot of points:
ax1.plot(uniform_numbers[0][mask_inside], uniform_numbers[1][mask_inside], '.')
ax1.plot(uniform_numbers[0][mask_outside], uniform_numbers[1][mask_outside], '.')
ax1.set(xlim=(-1, 1), ylim=(-1,1))

# Add circle to plot:
ax1.add_artist(plt.Circle((0, 0), 1, color='k', fill=False))


# plot a histogram of the distribution of pi estimates
Nbins2 = 10
xmin2 , xmax2 = 3, 3.3
binwidth2 = (xmax2 - xmin2) / Nbins2

fig2, ax2 = plt.subplots(figsize=(8, 8))
ax2.hist(pi_estimate, bins=Nbins2, range=(xmin2, xmax2))
ax2.set(xlim=(xmin2, xmax2), ylim=(0, 40))







if save_plots: 
    fig1.savefig("HitDist.pdf")
```
